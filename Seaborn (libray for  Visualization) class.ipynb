# Seaborn

#### seaborn is a python visualization library based on matplotlib that provides a high level interface for drawing attractive statistica graphics

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# sample data 
data=pd.DataFrame({
    'Category':['A','B','C','D','E'],
    'Value':[10,20,15,25,30]
})
```

# Bar Plot:
#### * Usefull for comparing categories or groups.
#### * Often used to visualize categorical data with numerical values.

```python
sns.barplot(x='Category',y='Value',data=data)
plt.title('Bar Plot')
plt.show()
```

# Scatter Plot:

####  ideal for visualizing the relationship between two continuous variables.
####  helps in identifying patterns,trends,or clusters in the data.

```python
# Scatter plot
sns.scatterplot(x='Category',y='Value',data=data,color='red')
plt.title('Scatter Plot')
plt.show()
```

# Line Plot

#### Effective for showing trends over time or continuous data.
#### Useful for time series analysis nd visualizing data with a sequential order.

```python
# line plot
sns.lineplot(x='Category',y='Value',data=data,marker='o')
plt.title('line Plot')
plt.show()
```

# Box Plot:

#### * Helpful in identifing outliers and understanding the distribution of numerical data.
#### * provides a visual summary of the central tendency,spread,and skewness of the data.

```python
import seaborn as sns
import pandas as pd

# sample data 
data=pd.DataFrame({
    'Category':['A','B','C','D','E'],
    'Value':[10,20,15,25,100]
})

# Box plot
sns.boxplot(x='Category',y='Value',data=data)
plt.title('Box Plot')
plt.show()
```

```python
# sample data 
data=pd.DataFrame({
    'Category':['A','B','C','D','E'],
    'Value':[10,20,15,25,30]
})

# Box plot
sns.boxplot(x='Category',y='Value',data=data)
plt.title('Box Plot')
plt.show()
```

# Histogram:

#### * Provides a visual representation of the distribution of a single numerical variable.
#### * Helps in understanding the shape,central tendency and variability of the data.

```python
# histogram
sns.histplot(data['Value'],bins=5,kde=True)
plt.title('Histogram')
plt.xlabel('Value')
plt.show()

# Kernel Density Estimation (KDE) is a technique used to estimate the probability density function (PDF) of a continuous random variable based on a finite set of data points. It provides a smooth estimate of the underlying distribution of the data.
```

# Heatmap:

#### * Useful for visualizing relationship or correlations in a matrix or tabular data.
#### * Often used in exploratory data analysis(EDA) and correlation analysis.

```python
# Heatmap (Requires a matrix)
import pandas as pd
import seaborn as sns

matrix_data = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})
```

```python
sns.heatmap(matrix_data,annot=True,cmap='YlGnBu')
plt.title('Heatmap')
plt.show()
# The annot argument in sns.heatmap() controls whether to display the numeric values in each cell of the heatmap.
# The cmap argument specifies the color map (colormap) to be used for coloring the heatmap cells.In this case, 'YlGnBu' represents a color map that transitions from yellow (Yl) to green (Gn) and blue (Bu).
```
