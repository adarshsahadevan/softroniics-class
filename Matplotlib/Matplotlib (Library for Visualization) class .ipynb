# 1.Line Plots
#### Line plots are common used to visualize the trend of a variable over a continuous interval.Matplotilb's plot function is the key to creating line plots.

```python
import matplotlib.pyplot as plt

# sample data
x=[1,2,3,4,5]
y=[2,4,6,8,10]

# Plotting a line graph
plt.plot(x,y, label='Line Plot')

#adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

#Display the legend
plt.legend()

#display the plot
plt.show() # in jupyter if we call plt it will get displayed, but in other platforms plt.show() should be called for display.
```

# 2.Scatter plot
#### scatter plots are effective for visualizing the relationship between two variables. Matplotib's scatter function is used for creating scatter plots.

```python
import matplotlib.pyplot as plt

# sample data
x=[1,2,3,4,5]
y=[2,4,6,8,10]

# Plotting a line graph
plt.scatter(x,y, label='Scatter plot',color='red',marker='+')

#adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

#Display the legend
plt.legend()

#display the plot
plt.show() 
```

# 3.Bar Plots
#### bar plots are used to compare categorical data.

```python
import matplotlib.pyplot as plt

# sample data
categories=['A','B','C','D','E']
values=[4,7,1,5,8]

# Plotting a line graph
plt.bar(categories,values, label='bar plot',color='green')

#adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('bar Plot Example')

#Display the legend
plt.legend()

#display the plot
plt.show() 
```

```python
import matplotlib.pyplot as plt

# sample data
categories=['A','B','C','D','E']
values=[4,7,1,5,8]

# Plotting a line graph
plt.barh(categories,values, label='bar plot',color='green')

#adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('bar Plot Example')

#Display the legend
plt.legend()

#display the plot
plt.show() 
```

# 4.Histograms

```python
import matplotlib.pyplot as plt
import numpy as np 

# generating random data for the histogram
data=np.random.randn(1000)

# Plotting a line graph
plt.hist(data,bins=50, edgecolor='black',alpha=0.8, label='Histogram',color='purple')
                             # alpha is the thikness of the color,edgecolor is the color of outerline of the bins. 
#adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('bar Plot Example')

#Display the legend
plt.legend()

#display the plot
plt.show() 
```

# Pie chart 

```python
import matplotlib.pyplot as plt
players='Rohit','Virat','Shikar','Yuvraj'
Runs=[45,30,15,10]
explode=(0,0.1,0,0) # it "explode" the 2nd slice

plt.pie(Runs,explode=explode,labels=players,autopct='%0.2f%%',shadow=True,startangle=180)

plt.show()
```

# figure class

#### Figure class is the top-leve; container one or more axes . It is the overall window or page on which everything is drawn.The figure keeps track of all the child axes, a group of 'special' Artists(titiles,figure legends,colorbars,etc),and even nested subfigures.

```python
# python program to show pyplot module
import matplotlib.pyplot as plt

# creating a new figure with width =5 inches
# and height =2 inches
fig=plt.figure(figsize=(4,2))

# creating a new axes for the figure
ax= fig.add_axes([1,1,1,1]) # [left,bottom,width,height]

x=[2,3,4,5,5,6,6,]
y=[5,7,1,3,4,6,8]

# adding the data to be plotted
ax.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
y=[1,4,9,16,25,36,49,64] #sales
x1=[1,16,30,42,55,68,77,88] #tv
x2=[1,6,12,18,28,40,52,65] #smartphone

fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

l1=ax.plot(x1,y,'ys-') # solid line with yellow colour and square marker
l2=ax.plot(x2,y,'go--') # dash line with green colour and circle marker

ax.legend(labels=('tv','smarphone'),loc='upper left')  # legend placed at lower right
ax.set_title("Advertisment effect on sales")
ax.set_xlabel("medium")
ax.set_ylabel('sales')
plt.show()
```

# Multiplots

## Method 1:Using the add_axes() method

```python
# python program to show pyplot module
import matplotlib.pyplot as plt

# creating a new figure with width =5 inches
# and height =4 inches
fig=plt.figure(figsize=(5,4))

# creating first axes for the figure
ax1=fig.add_axes([0.1,0.1,0.8,0.8])

# create second axes for the figure
ax2=fig.add_axes([0.5,0.5,0.3,0.3])

# adding the data to be plotted
ax1.plot([5,4,3,2,1],[2,3,4,5,6])
ax2.plot([1,2,3,4,5],[2,3,4,5,6])
plt.show()
```

# method 2: Using subplot() method.

```python
import matplotlib.pyplot as plt
# data to display on plots
x=[3,1,3]
y=[3,2,1]
z=[1,3,1]

#adding first subplot
plt.subplot(221) # no.of row 2,no.of coloumn 2,position of plot is 1
plt.plot(x,y)

#adding second subplot
plt.subplot(222)
plt.plot(z,y)

plt.subplot(223)
plt.plot(z,y)
```

# method 3: Using subplots() method.

```python
import matplotlib.pyplot as plt

# creating the figure and subplots
# according the argument passed
fig,axes=plt.subplots(1,2)  # 1 row and 2 coloumn

# plotting the data in the 1st subplot 
axes[0].plot([1,2,3,4],[1,2,3,4])

# plotting the data in the 1st subplot only
axes[0].plot([1,2,3,4],[4,3,2,1])

# plotting the data in the 2nd subplot only 
axes[1].plot([1,2,3,4],[1,1,1,1])
```

# Setting Limits and Tick Labels

```python
# python program to show pyplot module
import matplotlib.pyplot as plt

x=[3,1,3]
y=[3,2,1]

# creating a new figure with width =5 inches
# and height =4 inches
fig=plt.figure(figsize=(5,4))

# creating first axes for the figure
ax=fig.add_axes([0.1,0.1,0.8,0.8])

# adding the data to be plotted
ax.plot(x,y)
ax.set_xlim(1,2) # limit of x axis
ax.set_xticklabels(("one","two","three","four","five","six"))
plt.show()
```

# Annotations

```python
import numpy as np
fig,ax=plt.subplots(figsize=(5,2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)

line=ax.plot(t,s,lw=2) # lw is line width

ax.annotate('local max', xy=(2,1),xytext=(3,1.5),
           arrowprops=dict(facecolor='black',shrink=0.01))

ax.set_ylim(-2,2);
```

# Area plots 

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]

sleeping =[7,8,6,11,7]
eating =[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('stack plot')
plt.legend()
plt.show()
```
